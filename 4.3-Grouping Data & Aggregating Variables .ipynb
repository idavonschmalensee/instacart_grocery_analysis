# Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.


```python
#New column was created by the grouped variables
ords_prods_merge['max'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
pd.options.display.max_rows = None
```

```python
# Flags were created by loc()
ords_prods_merge.loc[ords_prods_merge['max'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max'] <= 40) & (ords_prods_merge['max'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# There are more regular customers than loyal or new customers
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

# If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

# If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# #New column was created by the grouped variables
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.me)
```

```python
# flags were assigned to variables
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price']>=10,'spending_flag']='High spender'
```

```python
ords_prods_merge.head()
```

```python
# There are more low spender than high spender. Customers tend to not to spend much in one time.
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
# If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#New column was created by the grouped variables
ords_prods_merge['median_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# flags were assigned
ords_prods_merge.loc[ords_prods_merge['median_order']>20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_order']<=10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_order'] > 10) & (ords_prods_merge['median_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.head()
```

```python
# NaN are the new customers that are not frequent or non-frequent yet. There are more frequent customers than non-frequent
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
